```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
import torch
import torch.nn as nn
from torch.optim import Adam
```

```python
# load dataset
file_path = 'Fuel_Consumption_2000-2022.csv'
data = pd.read_csv(file_path)
data = data[['YEAR', 'COMB (L/100 km)']]
data['YEAR'] = pd.to_datetime(data['YEAR'], format='%Y')
data.set_index('YEAR', inplace=True)
```

```python
# Resemple annually
annual_date = data.resample('YE').mean()
scaler = MinMaxScaler()
annual_data['Scaled_Comb'] = scaler.fit_transform(annual_data[['COMB (L/100 km)']])
```

```python
# plot ACF and PACF
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(annual_data['COMB (L/100 km)'])
plt.title('ACF - Fuel Consumption')
plt.show()
plot_pacf(annual_data['COMB (L/100 km)'])
plt.title('PACF - Fuel Consumption')
plt.show()
```

```python
# Train-Test split
train_size = int(len(annual_data) * 0.67)
train, test = annual_data.iloc[:train_size], annual_data.iloc[train_size:]
```

```python
# -- ARIMA -- 
model_arima = ARIMA(train['COMB (L/100 km)'], order=(1, 1, 1))
results_arima = model_arima.fit()
forecast_arima = results_arima.get_forecast(steps=len(test))
arima_pred = forecast_arima.predicted_mean.values
true_values_arima = test['COMB (L/100 km)'].values
```

```python
# -- LSTM --
train_tensor = torch.tensor(train[['Scaled_Comb']].values, dtype=torch.float32).view(-1, 1, 1)
test_tensor = torch.tensor(test[['Scaled_Comb']].values, dtype=torch.float32).view(-1, 1, 1)
```

```python
class LSTMModel(nn.Module):
    def __init__(self):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size=1, hidden_size=100, batch_first=True)
        self.linear = nn.Linear(100, 1)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.linear(lstm_out[:, -1, :])

model_lstm = LSTMModel()
loss_fn = nn.MSELoss()
optimizer = Adam(model_lstm.parameters(), lr=0.01)
for epoch in range(100):
    model_lstm.train()
    optimizer.zero_grad()
    output = model_lstm(train_tensor)
    loss = loss_fn(output, train_tensor[:, -1, :])
    loss.backward()
    optimizer.step()

model_lstm.eval()
with torch.no_grad():
    lstm_output = model_lstm(test_tensor)
    lstm_pred = scaler.inverse_transform(lstm_output.numpy())
```

```python
# -- Lag-based ML models --
annual_data['Lag1'] = annual_data['Scaled_Comb'].shift(1)
ml_data = annual_data.dropna()
X = ml_data[['Lag1']]
y = ml_data['Scaled_Comb']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=len(test)/len(ml_data), shuffle=False)
```

```python
# Add ARIMA and LSTM manually
metrics["ARIMA"] = [
    np.sqrt(mean_squared_error(true_values_arima, arima_pred)),
    mean_absolute_error(true_values_arima, arima_pred),
    r2_score(true_values_arima, arima_pred)
]

metrics["LSTM"] = [
    np.sqrt(mean_squared_error(true_values_arima, lstm_pred)),
    mean_absolute_error(true_values_arima, lstm_pred),
    r2_score(true_values_arima, lstm_pred)
]
```

```python
# format results
results_df = pd.DataFrame(metrics, index=['RMSE', 'MAE', 'R²']).T.reset_index().rename(columns={'index': 'Model'})
```

```python
# display results table
print("Model Evaluation Results:\n")
# header
print("{:<18} {:>10} {:>10} {:>12}".format("Model", "RMSE", "MAE", "R²"))
print("-" * 52)

# rows
for index, row in results_df.iterrows():
    print("{:<18} {:>10.6f} {:>10.6f} {:>12.6f}".format(
        row['Model'], row['RMSE'], row['MAE'], row['R²']
    ))
```
